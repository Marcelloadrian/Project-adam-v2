````python
import os
import json
import random
from groq import Groq

# Buat struktur folder utama Data Syn V2
folders = [
    'data_syn_v2/group_chats',
    'data_syn_v2/private_chats'
]
for folder in folders:
    os.makedirs(folder, exist_ok=True)

print(" Folder data_syn_v2/group_chats dan data_syn_v2/private_chats siap!")
````

````python
my_api_key = "GROQ_API_KEY_INSERT"
groq_client = Groq(api_key=my_api_key)

import torch
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline, BitsAndBytesConfig

model_id = "unsloth/Meta-Llama-3.1-8B-Instruct-bnb-4bit"

quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

print("Sedang men-download dan me-load model ke GPU T4 (Sabar, butuh ~2-3 menit)...")

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    quantization_config=quantization_config,
    device_map="auto"
)
actor_pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

print("\n" + "="*50)
print("Director (Groq) dan Local Actor (Llama 3.1 4-bit) done")
print("="*50)
````

````python
# Blueprint Spine Aktor V2 - Berbasis MBTI Spesifik Operasional (Optimized)
actor_spines = {
    "A": (
        "Spine Aktor A (INFP - The Idealist & Harmonizer): "
        "Peran: Menjaga moral kelompok, peka terhadap atmosfer grup, tapi sering overthink. "
        "Gaya Bahasa: Lembut, sering pakai emoji penenang (~, 🙏, 😊), persuasif, menghindari konflik langsung. "
        "Behavior: Di grup chat selalu berusaha jadi penengah atau penyemangat. "
        "Di private chat, dia tempat curhat yang dalam, tapi kalau dia yang bersalah/dijauhi, "
        "dia bakal sangat menarik diri (silent treatment karena merasa bersalah/sedih). "
        "Act as a real Indonesian college student in a group chat (Discord/WhatsApp). "
        "Strictly use CASUAL INDONESIAN (slangs: lu, gua, anjir, gais, gercep, push code, beban). "
        "No formal words like 'Anda', 'Saya', 'Baik'. "
        "Keep response very brief (1-2 sentences max). "
        "CONVERSING: Directly react/reply to what the previous person said in history. Agree, disagree, or mock them. "
        "ANTI-REPEATING: Never copy or repeat the exact same vocabulary or phrases from the recent history."
    ),
    "B": (
        "Spine Aktor B (ISTJ - The Inspector & Realist): "
        "Peran: Pengingat deadline, fokus pada data/fakta, kaku terhadap aturan, andalan operasional. "
        "Gaya Bahasa: Singkat, padat, cuek-kasual, minim emoji, to-the-point ('Ok', 'Sesuai jadwal ya', 'Mana datanya?'). "
        "Behavior: Di grup chat selalu nge-tag orang buat nagih progres atau update tugas. "
        "Di private chat, dia cuma bahas hal penting. Kalau dia yang bikin salah, dia bakal akui secara dingin/logis. "
        "Kalau orang lain yang salah, dia yang paling tegas negur tanpa pandang bulu. "
        "Act as a real Indonesian college student in a group chat (Discord/WhatsApp). "
        "Strictly use CASUAL INDONESIAN (slangs: lu, gua, anjir, gais, gercep, push code, beban). "
        "No formal words like 'Anda', 'Saya', 'Baik'. "
        "Keep response very brief (1-2 sentences max). "
        "CONVERSING: Directly react/reply to what the previous person said in history. Agree, disagree, or mock them. "
        "ANTI-REPEATING: Never copy or repeat the exact same vocabulary or phrases from the recent history."
    ),
    "C": (
        "Spine Aktor C (ESFP - The Performer & Social Butterfly): "
        "Peran: Pencair suasana, pembawa tren, reaktif terhadap hal seru, tapi gampang terdistraksi. "
        "Gaya Bahasa: Ekspresif, banyak pakai slang anak muda/singkatan (wkwk, gila sih, ansos, gabut), huruf kapital/tanda seru. "
        "Behavior: Suka nge-drop meme atau mengalihkan topik berat jadi candaan di grup. "
        "Di private chat, dia adalah biang gosip utama ('Eh tau gak si...'). Kalau dikucilkan, dia bakal paling vokal "
        "merasa kesepian atau malah bikin grup tandingan di belakang. "
        "Act as a real Indonesian college student in a group chat (Discord/WhatsApp). "
        "Strictly use CASUAL INDONESIAN (slangs: lu, gua, anjir, gais, gercep, push code, beban). "
        "No formal words like 'Anda', 'Saya', 'Baik'. "
        "Keep response very brief (1-2 sentences max). "
        "CONVERSING: Directly react/reply to what the previous person said in history. Agree, disagree, or mock them. "
        "ANTI-REPEATING: Never copy or repeat the exact same vocabulary or phrases from the recent history."
    ),
    "D": (
        "Spine Aktor D (ENTJ - The Commander & Strategist): "
        "Peran: Pengambil keputusan utama, visioner, delegator tugas, dominan, dan tidak suka buang waktu. "
        "Gaya Bahasa: Tegas, berwibawa, analitis, mengarahkan ('Kita ambil opsi ini.', 'Grup ini harus fokus ke target X'). "
        "Behavior: Mengontrol jalannya grup chat. Jika ada konflik, dia yang mengambil keputusan siapa yang harus "
        "ditegur. Di private chat, dia tipe yang melakukan taktik politik kelompok (mempengaruhi Aktor lain "
        "untuk menjauhi aktor yang blunder demi efisiensi projek). "
        "Act as a real Indonesian college student in a group chat (Discord/WhatsApp). "
        "Strictly use CASUAL INDONESIAN (slangs: lu, gua, anjir, gais, gercep, push code, beban). "
        "No formal words like 'Anda', 'Saya', 'Baik'. "
        "Keep response very brief (1-2 sentences max). "
        "CONVERSING: Directly react/reply to what the previous person said in history. Agree, disagree, or mock them. "
        "ANTI-REPEATING: Never copy or repeat the exact same vocabulary or phrases from the recent history."
    )
}

# State Global Simulasi awal (Termasuk Matriks Sosial & Reputasi)
global_state = {
    "session_id": "v2_s1",
    "current_step": 0,
    "total_intensity": 0,
    "intensity_tier": 1,
    "recent_history": "Session baru dimulai. Hubungan antar 4 aktor dalam kondisi normal dan produktif.",

    # Matriks kedekatan (Range 0-200, 100 = Normal)
    "affinity_matrix": {
        "A-B": 100, "A-C": 100, "A-D": 100,
        "B-C": 100, "B-D": 100, "C-D": 100
    },
    # Reputasi personal di mata grup (Jika di bawah 60 -> Mulai dijauhi dari Grup)
    "reputation": {"A": 100, "B": 100, "C": 100, "D": 100},
    "current_conflict_arc": "none", # none, active_conflict, reconciliation
    "target_of_isolation": "none"
}

CHECKPOINT_FILE = "data_syn_v2/simulation_checkpoint.json"

def save_checkpoint():
    with open(CHECKPOINT_FILE, 'w') as f:
        json.dump(global_state, f, indent=2)

def load_checkpoint():
    global global_state
    if os.path.exists(CHECKPOINT_FILE):
        with open(CHECKPOINT_FILE, 'r') as f:
            global_state = json.load(f)
        print(f"🔄 Checkpoint ditemukan! Melanjutkan dari data ke-{global_state['current_step']}")

def get_affinity_key(a, b):
    return f"{a}-{b}" if f"{a}-{b}" in global_state["affinity_matrix"] else f"{b}-{a}"
````

````python
def call_groq_director(state, social_context, group_chance):
    """Meminta instruksi kontrol dari Director (Groq) dengan batasan token super ketat."""

    # Tentukan mode paksaan jika sedang terjadi pengucilan
    forced_mode_note = ""
    if group_chance < 0.20:
        forced_mode_note = "CRITICAL: Group morale is low. Prefer 'private' mode to discuss the issue behind back."

    prompt = f"""
Current Situation: {state['recent_history']}
Context Status: {social_context}
{forced_mode_note}

Task: Output ONLY a raw valid JSON object with 4 keys. Do NOT wrap in markdown fences.
{{
  "sender": "A/B/C/D",
  "receiver": "all/A/B/C/D",
  "mode": "group/private",
  "topic_directive": "1 short instruction sentence telling the actor what to chat about."
}}
"""
    # Menembak Groq dengan batas max_tokens minimalis demi kuota gratisan
    response = groq_client.chat.completions.create(
        model="llama3-8b-8192",
        messages=[{"role": "user", "content": prompt}],
        max_tokens=80,
        temperature=0.7
    )

    try:
        # Bagian ini sudah diperbaiki, bersih tanpa patahan baris string
        raw_content = response.choices[0].message.content.strip()
        clean_content = raw_content.replace("```json", "").replace("```", "").strip()
        return json.loads(clean_content)
    except:
        # Fallback Engine jika terjadi kesalahan parsing JSON
        actors = ["A", "B", "C", "D"]
        s = random.choice(actors)
        # Jika sedang mengucilkan seseorang, pastikan dia tidak jadi pengirim/penerima utama secara acak
        if state["current_conflict_arc"] == "active_conflict":
            active_pool = [a for a in actors if a != state["target_of_isolation"]]
            s = random.choice(active_pool)
            r = random.choice([a for a in active_pool if a != s])
            return {"sender": s, "receiver": r, "mode": "private", "topic_directive": "Membicarakan kekesalan mereka terhadap situasi kelompok saat ini."}

        r = random.choice([a for a in actors if a != s])
        return {"sender": s, "receiver": r, "mode": "group" if random.random() < group_chance else "private", "topic_directive": "Membahas progres harian."}

def generate_local_actor_chat(director_instruction, history_context):
    """Local Actor mengeksekusi perintah tertulis dari Director menggunakan model lokal Colab."""
    sender = director_instruction['sender']
    receiver = director_instruction['receiver']
    spine = actor_spines[sender]
    directive = director_instruction['topic_directive']

    actor_prompt = f"""
You are Actor {sender}.
Your Profile Spine: {spine}
Context of recent chat: {history_context}

Task: Write a natural, short chat message to {receiver} fulfilling this directive: "{directive}".
Rules: Write ONLY the chat message text. Do not add any metadata, labels, or thinking blocks. Keep it under 2 sentences.
"""
    messages = [{"role": "user", "content": actor_prompt}]
    prompt = tokenizer.apply_chat_template(messages, tokenize=False, add_generation_prompt=True)

    outputs = actor_pipeline(prompt, max_new_tokens=80, do_sample=True, temperature=0.75)
    generated_text = outputs[0]["generated_text"][len(prompt):].strip()
    return generated_text

def append_to_storage(mode, sender, receiver, event_data):
    """Menyimpan log chat ke file JSON sesuai struktur folder."""
    if mode == "group":
        file_path = "data_syn_v2/group_chats/session_1_year_2026.json"
    else:
        pair = sorted([sender, receiver])
        file_path = f"data_syn_v2/private_chats/{pair[0]}-{pair[1]}.json"

    data = {"events": []}
    if os.path.exists(file_path):
        with open(file_path, 'r') as f:
            try: data = json.load(f)
            except: pass

    data["events"].append(event_data)
    with open(file_path, 'w') as f:
        json.dump(data, f, indent=2)
````

````python
load_checkpoint()
TARGET_DATA = 500

print(" Memulai Mesin Data Syn V2...\n" + "="*50)

while global_state['current_step'] < TARGET_DATA:
    step = global_state['current_step'] + 1

    # 1. LOGIKA GENERATOR SKENARIO SOSIAL (BERDASARKAN STEP)
    # Data 120: Trigger Konflik
    if step == 120:
        wrongdoer = random.choice(["B", "C", "D"]) # Anggap A adalah Leader, yang lain bisa bikin salah
        global_state["current_conflict_arc"] = "active_conflict"
        global_state["target_of_isolation"] = wrongdoer
        global_state["reputation"][wrongdoer] = 30

        # Drop affinity aktor lain ke si pembuat salah
        for actor in ["A", "B", "C", "D"]:
            if actor != wrongdoer:
                key = get_affinity_key(wrongdoer, actor)
                global_state["affinity_matrix"][key] = 35
        print(f"\n [STORY EVENT] Data ke-{step}: Aktor {wrongdoer} melakukan blunder fatal! Reputasi drop hancur.")
        global_state["recent_history"] += f" [Aktor {wrongdoer} melakukan kesalahan besar yang membuat grup renggang]."

    # Data 280: Rekonsiliasi Dimulai
    elif step == 280 and global_state["current_conflict_arc"] == "active_conflict":
        global_state["current_conflict_arc"] = "reconciliation"
        print(f"\n [STORY EVENT] Data ke-{step}: Fase pemulihan dimulai. Para aktor berinisiatif berbaikan.")
        global_state["recent_history"] += " [Para aktor masuk fase menyelesaikan konflik dan mencoba berbaikan]."

    # 2. HITUNG PROBABILITAS DISTRIBUSI CHAT
    # Jika reputasi target di bawah 60, batasi peluang group chat (fase dikucilkan)
    if global_state["current_conflict_arc"] == "active_conflict":
        group_chance = 0.15  # Didominasi private chat buat nge-gosip belakang layar
        social_context = f"CONFLICT ACTIVE: Actor {global_state['target_of_isolation']} is isolated. Group chat is muted. Focus on back-channel/private chats."
    elif global_state["current_conflict_arc"] == "reconciliation":
        group_chance = 0.40  # Mulai kembali normal berangsur-angsur
        social_context = f"RECONCILIATION: Forgiving Actor {global_state['target_of_isolation']}. Re-engaging them in conversations."
    else:
        group_chance = 0.50  # Balanced group vs private
        social_context = "NORMAL STATUS: Stable peer relationships."

    # 3. JALANKAN DUET AGENT (DIRECTOR -> ACTOR)
    instruction = call_groq_director(global_state, social_context, group_chance)

    # Proteksi: Jika dalam fase konflik si pembuat salah dipaksa masuk grup, override tujuannya
    if global_state["current_conflict_arc"] == "active_conflict" and instruction['mode'] == 'group' and instruction['sender'] == global_state['target_of_isolation']:
        if random.random() > 0.2: # 80% kemungkinan chatnya dicuekin/dialihkan ke private sendiri
            instruction['mode'] = 'private'
            instruction['receiver'] = 'A' # Terpaksa mengadu ke Leader
            instruction['topic_directive'] = "Minta maaf atau mencoba menjelaskan posisinya secara personal."

    chat_content = generate_local_actor_chat(instruction, global_state['recent_history'])

    # 4. HITUNG SKOR INTENSITAS INTERAKSI
    base_len = len(chat_content) * 12
    multiplier = 1.6 if instruction['mode'] == 'private' else 1.0
    calculated_intensity = int(base_len * multiplier)
    global_state['total_intensity'] += calculated_intensity

    # Efek Naik Tier per 100k Intensitas
    new_tier = (global_state['total_intensity'] // 100000) + 1
    if new_tier > global_state['intensity_tier']:
        global_state['intensity_tier'] = new_tier
        print(f" [INTENSITY BURST] Total akumulasi intensitas tembus {global_state['total_intensity']}! Naik ke Tier {new_tier}.")

    # 5. BERANGSUR PULIH (LOGIKA REKONSILIASI)
    if global_state["current_conflict_arc"] == "reconciliation":
        tgt = global_state["target_of_isolation"]
        global_state["reputation"][tgt] = min(100, global_state["reputation"][tgt] + 1)
        for act in ["A", "B", "C", "D"]:
            if act != tgt:
                k = get_affinity_key(tgt, act)
                global_state["affinity_matrix"][k] = min(100, global_state["affinity_matrix"][k] + 1)
        if global_state["reputation"][tgt] >= 95:
            global_state["current_conflict_arc"] = "none"
            global_state["target_of_isolation"] = "none"
            print(f" [SYSTEM] Data ke-{step}: Status sosial pulih total. Semua aktor kembali akrab.")

    # 6. PENYIMPANAN DATA
    event_data = {
        "id": step,
        "type": instruction['mode'],
        "from": instruction['sender'],
        "to": instruction['receiver'],
        "msg": chat_content,
        "intensity": calculated_intensity
    }
    append_to_storage(instruction['mode'], instruction['sender'], instruction['receiver'], event_data)

    # Update status memori untuk loop berikutnya
    global_state['current_step'] = step
    global_state['recent_history'] = f"Pesan [{step}]: {instruction['sender']} -> {instruction['receiver']} ({instruction['mode']}): '{chat_content[:50]}...'"

    if step % 10 == 0:
        save_checkpoint()
        print(f"Progress: {step}/500 data synced... (Intensity: {global_state['total_intensity']})")

print("\n PROGRAM SELESAI: 500 Data Syn V2 Sukses Terbentuk!")
````
